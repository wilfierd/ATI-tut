## Exercise 3.2 Specifying a model


In this exercise, you will work with a .csv file that contains some data about workers in a company. You can download this file from the following link:

[hourly_wages.csv](https://drive.google.com/file/d/1elGBb48CX728knwJzU3ImHNvtuRoG7vu/view?usp=drive_link)

The data is pre-loaded into a pandas DataFrame called `df`.

In this exercise, you'll create a neural network with two hidden layers and an output layer.

The target variable you'll be predicting is `wage_per_hour`. For convenience, everything in `df` except for the target has been converted to a NumPy array called `predictors`. The target, `wage_per_hour`, is available as a NumPy array called `target`.

The network can be created with the `Sequential` model constructor and the `Dense` layer constructor.

## Instructions

* Store the number of columns in the `predictors` data to `n_cols`. This has been done for you.
* Start by creating a `Sequential` model called model.
* Use the `add()` method on model to add a `Dense` layer as the first hidden layer of the network.
  * Add `50` units, specify `activation='relu'`, and the `input_shape` parameter to be the tuple `(n_cols,)` which means it has `n_cols` items in each row of data, and any number of rows of data are acceptable as inputs.
* Add another `Dense` layer. This should have `32` units and a `'relu'` activation.
* Add an output layer, which is a `Dense` layer with a single node. Don't use any activation function here.
* Finally, call model.summary() to print a useful summary of the model

## Code

Load data and convert the data to NumPy array:

```python
import pandas as pd
# Load csv file into the dataframe: df
df = pd.read_csv("hourly_wages.csv")
# Split the dataframe df into two dataframes:
wagePerHourDf = df.iloc[:,0]
predictorsDf = df.iloc[:,1:df.shape[1]]

# Create predictors NumPy array: predictors
predictors = predictorsDf.to_numpy()

# Create target NumPy array: target
target = wagePerHourDf.to_numpy()
```

Create the neural network

```python
# Import necessary modules
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

# Save the number of columns in predictors: n_cols
n_cols = predictors.shape[1]

# Set up the model: model
model = Sequential()

# Add the first layer
model.add(Dense(50, activation='relu', input_shape=(n_cols,)))

# Add the second layer
model.add(Dense(32, activation='relu'))

# Add the output layer
model.add(Dense(1))

model.summary()
```
